```python
from data_generation import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
days = 365
init_S, init_I, init_R = 100, 2, 0
beta, gamma = 0.3, 0.1
```

```python
initial_conditions = [init_S, init_I, init_R]
tspan = np.arange(0, days, 1)
```

```python
sol = ode_solver(initial_conditions, tspan, beta, gamma)
```

```python
S, I, R = sol.T
```

```python
plt.figure(figsize=(12, 6))
plt.plot(tspan, S, label='Susceptible')
plt.plot(tspan, I, label='Infected')
plt.plot(tspan, R, label='Recovered')
plt.xlabel('Time (days)')
plt.ylabel('Proportion of Population')
plt.title('SIR Model Simulation')
plt.legend()
plt.tight_layout()
plt.show()
```
